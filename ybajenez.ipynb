```python
import numpy as np

#Question 1: Analyzing Monthly Expenses
#a.
# Expenses data for the current and previous year
current_year_expenses = [1200, 1350, 1100, 980, 1500]
previous_year_expenses = [1150, 1300, 1050, 950, 1400]

# Now converting the expense lists to NumPy arrays
current_year_array = np.array(current_year_expenses)
previous_year_array = np.array(previous_year_expenses)

# Calculating statistics for the current year's expenses
current_mean = np.mean(current_year_array)
current_median = np.median(current_year_array)
current_std_dev = np.std(current_year_array)
current_total = np.sum(current_year_array)

# Calculating statistics for the previous year's expenses
previous_mean = np.mean(previous_year_array)
previous_median = np.median(previous_year_array)
previous_std_dev = np.std(previous_year_array)
previous_total = np.sum(previous_year_array)

# Printing the statistics for both years
print("Statistics for Current Year's Expenses:")
print(f"Mean Expense: {current_mean}")
print(f"Median Expense: {current_median}")
print(f"Standard Deviation of Expenses: {current_std_dev}")
print(f"Total Expenses: {current_total}\n")

print("Statistics for Previous Year's Expenses:")
print(f"Mean Expense: {previous_mean}")
print(f"Median Expense: {previous_median}")
print(f"Standard Deviation of Expenses: {previous_std_dev}")
print(f"Total Expenses: {previous_total}")
```

```python
#b.
# Calculating the difference and percentage change in mean expenses
difference_mean_expenses = current_mean - previous_mean
percentage_change = ((current_mean - previous_mean) / previous_mean) * 100

# Calculating coefficient of variation (CV) for both years
current_cv = (current_std_dev / current_mean) * 100
previous_cv = (previous_std_dev / previous_mean) * 100

# Printing the difference and percentage change
print("\nDifference between the mean expenses of the current year and the previous year:", difference_mean_expenses)
print("Percentage change in mean expenses from previous year to current year:", percentage_change, "%")

```

```python
#C.
# Function to determine risk assessment based on the coefficient of variation
def risk_assessment(cv):
    if cv < 20:
        return "Low Risk"
    elif 20 <= cv < 50:
        return "Moderate Risk"
    else:
        return "High Risk"

# Geting the risk assessment messages for both years
current_year_risk = risk_assessment(current_cv)
previous_year_risk = risk_assessment(previous_cv)

# Printing the risk assessment metrics for both years
print("\nRisk Assessment Metrics:")
print("Current Year - Coefficient of Variation (CV):", current_cv)
print("Risk Assessment for Current Year:", current_year_risk)

print("\nPrevious Year - Coefficient of Variation (CV):", previous_cv)
print("Risk Assessment for Previous Year:", previous_year_risk)
```

```python
def identify_high_low_sales(months, sales):
    # Create a dictionary pairing each month with its sales
    monthly_sales = dict(zip(months, sales))

    # Identify the month with the highest sales
    max_sales_month = max(monthly_sales, key=monthly_sales.get)
    max_sales_value = monthly_sales[max_sales_month]

    # Identify the month with the lowest sales
    min_sales_month = min(monthly_sales, key=monthly_sales.get)
    min_sales_value = monthly_sales[min_sales_month]

    return max_sales_month, max_sales_value, min_sales_month, min_sales_value

# Identifying months with highest and lowest sales
max_sales_month, max_sales_value, min_sales_month, min_sales_value = identify_high_low_sales(months, sales)

print(f"Month with the highest sales: {max_sales_month} with {max_sales_value} sales")
print(f"Month with the lowest sales: {min_sales_month} with {min_sales_value} sales")
```
