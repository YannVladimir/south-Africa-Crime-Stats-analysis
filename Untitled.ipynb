```python
import numpy as np

# Expenses data for the current and previous year
current_year_expenses = [1200, 1350, 1100, 980, 1500]
previous_year_expenses = [1150, 1300, 1050, 950, 1400]

# Convert the expense lists to NumPy arrays
current_year_array = np.array(current_year_expenses)
previous_year_array = np.array(previous_year_expenses)

# Calculate statistics for the current year's expenses
current_mean = np.mean(current_year_array)
current_median = np.median(current_year_array)
current_std_dev = np.std(current_year_array)
current_total = np.sum(current_year_array)

# Calculate statistics for the previous year's expenses
previous_mean = np.mean(previous_year_array)
previous_median = np.median(previous_year_array)
previous_std_dev = np.std(previous_year_array)
previous_total = np.sum(previous_year_array)

# Print the statistics for both years
print("Statistics for Current Year's Expenses:")
print(f"Mean Expense: {current_mean}")
print(f"Median Expense: {current_median}")
print(f"Standard Deviation of Expenses: {current_std_dev}")
print(f"Total Expenses: {current_total}\n")

print("Statistics for Previous Year's Expenses:")
print(f"Mean Expense: {previous_mean}")
print(f"Median Expense: {previous_median}")
print(f"Standard Deviation of Expenses: {previous_std_dev}")
print(f"Total Expenses: {previous_total}")
```

```python

# Calculate the difference and percentage change in mean expenses
difference_mean_expenses = current_mean - previous_mean
percentage_change = ((current_mean - previous_mean) / previous_mean) * 100

# Calculate coefficient of variation (CV) for both years
current_cv = (current_std_dev / current_mean) * 100
previous_cv = (previous_std_dev / previous_mean) * 100

# Print the difference and percentage change
print("\nDifference between the mean expenses of the current year and the previous year:", difference_mean_expenses)
print("Percentage change in mean expenses from previous year to current year:", percentage_change, "%")

```

```python

# Function to determine risk assessment based on the coefficient of variation
def risk_assessment(cv):
    if cv < 20:
        return "Low Risk"
    elif 20 <= cv < 50:
        return "Moderate Risk"
    else:
        return "High Risk"

# Get risk assessment messages for both years
current_year_risk = risk_assessment(current_cv)
previous_year_risk = risk_assessment(previous_cv)

# Print the risk assessment metrics for both years
print("\nRisk Assessment Metrics:")
print("Current Year - Coefficient of Variation (CV):", current_cv)
print("Risk Assessment for Current Year:", current_year_risk)

print("\nPrevious Year - Coefficient of Variation (CV):", previous_cv)
print("Risk Assessment for Previous Year:", previous_year_risk)
```

```python
# Now 2nd Question

# Historical sales data for the last three years
year1_sales = [12000, 13500, 11000, 9800, 15000, 12300, 13700, 11150, 9750, 14850, 13000, 15500]
year2_sales = [12500, 13000, 11200, 9900, 15200, 12600, 14200, 11350, 9950, 15000, 13500, 16000]
year3_sales = [12300, 13700, 11150, 9750, 14850, 12800, 13900, 11000, 9600, 14500, 12750, 14200]

# Convert the sales data lists to NumPy arrays
year1_array = np.array(year1_sales)
year2_array = np.array(year2_sales)
year3_array = np.array(year3_sales)

# Calculate the monthly average (mean) sales across the last three years
average_monthly_sales = np.mean([year1_array, year2_array, year3_array], axis=0)

# Print the monthly sales forecast for the upcoming year
print("Monthly Sales Forecast for the Upcoming Year:")
for month, sales in enumerate(average_monthly_sales, start=1):
    print(f"Month {month}: {sales:.2f}")
```

```python
# Generating X values for linear regression
x_values = np.arange(len(months))

# Perform linear regression (fit a line to the data)
slope, intercept = np.polyfit(x_values, sales, 1)
trendline = slope * x_values + intercept

# Plotting the line graph with markers and trendline
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', linestyle='--', color='g', label='Sales')  # Data points
plt.plot(months, trendline, linestyle='-', label='Trendline')  # Trendline

# Adding title and labels
plt.title('Monthly Sales Data with Trendline')
plt.xlabel('Months')
plt.ylabel('Sales')

# Adding labels for data points
for i, sale in enumerate(sales):
    plt.text(months[i], sale, sale, ha='right', va='bottom')

# Adding labels for the trendline
for i, trend in enumerate(trendline):
    plt.text(months[i], trend, f'{trend:.2f}', ha='left', va='top')

# Adding a legend
plt.legend()

# Show the plot
plt.grid(True, linestyle='--')
plt.show()
```

```python
def identify_high_low_sales(months, sales):
    # Create a dictionary pairing each month with its sales
    monthly_sales = dict(zip(months, sales))

    # Identify the month with the highest sales
    max_sales_month = max(monthly_sales, key=monthly_sales.get)
    max_sales_value = monthly_sales[max_sales_month]

    # Identify the month with the lowest sales
    min_sales_month = min(monthly_sales, key=monthly_sales.get)
    min_sales_value = monthly_sales[min_sales_month]

    return max_sales_month, max_sales_value, min_sales_month, min_sales_value

# Identifying months with highest and lowest sales
max_sales_month, max_sales_value, min_sales_month, min_sales_value = identify_high_low_sales(months, sales)

print(f"Month with the highest sales: {max_sales_month} with {max_sales_value} sales")
print(f"Month with the lowest sales: {min_sales_month} with {min_sales_value} sales")
```

```python
# Monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [10000, 12000, 9000, 11000, 13000]

# Special events or promotions
special_events = {
    'Feb': 'Product Launch',  # Example event in February
    'Apr': 'Spring Sale'      # Example event in April
}

# Plotting the line graph
plt.figure(figsize=(10, 6))

# Plotting sales data
plt.plot(months, sales, marker='o', linestyle='-', color='g', label='Monthly Sales')

# Marking special events
for month, event in special_events.items():
    index = months.index(month)
    plt.scatter(index, sales[index], color='red', marker='^', s=100, label=f'{event} ({month})')
    plt.text(index, sales[index], f'{event}\n({month})', ha='right', va='bottom')

# Adding title and labels
plt.title('Monthly Sales with Highlighted Events')
plt.xlabel('Months')
plt.ylabel('Sales')

# Adding a legend
plt.legend()

# Show the plot
plt.grid(True, linestyle='--')
plt.show()
```
